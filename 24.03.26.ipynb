```python

with_dash = "The transaction occurred on 24-03-2026 at 16:45."
Without_dash = re.sub(r'(\d{2})-(\d{2})-(\d{4}) at (\d{2}):(\d{2})', r'\1.\2.\3 at \4hrs:\5mins', with_dash)
print(f"Original: {with_dash}")
print(f"Cleaned:  {Without_dash}")
```

```python
clean_date = "Processing Order: 98765 for the 22.03.2026 shipment. cost is 20"
final_output = re.sub(r'\w+:\s*(\d+)', r'REF#\1', clean_date)
print(f"Before: {clean_date}")
print(f"After:  {final_output}")
```

Start,^,No text allowed before the email.
User,[\w.%+-]+,Matches the account name.
At,@,The mandatory separator.
Host,[\w.-]+,Matches the website/server name.
Dot,\.,The literal period before the extension.
Ext,"[A-Z]{2,}","Matches .com, .net, etc. (min 2 letters)."
End,$,No text allowed after the email.

```python
import re

def email_validator(email):
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    
    if re.match(pattern, email):
        print(f"{email} is a valid email format.")
    else:
        print(f"{email} is NOT a valid email format.")

email_validator("user.name@domain.com") 
email_validator("test-123@sub.school.edu")
email_validator("invalid-email@domain")
```
