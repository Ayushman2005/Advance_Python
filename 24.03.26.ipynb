Start Anchor,^,Ensures the string must start with the email (no spaces or text before it).
Username,[\w\.-]+,"Matches the name. \w (letters/numbers), \. (dot), and - (dash). The + means ""one or more."""
The Divider,@,The mandatory separator between the user and the host.
Domain,[\w\.-]+,"Matches the provider (e.g., gmail, outlook, or my.school)."
The Extension,\.\w+,Matches a literal dot \. followed by the suffix like com or org.
End Anchor,$,Ensures the string must end exactly after the extension (no text after it).

```python
import re

def email_validator(email):
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    
    if re.match(pattern, email):
        print(f"{email} is a valid email format.")
    else:
        print(f"{email} is NOT a valid email format.")

email_validator("user.name@domain.com") 
email_validator("test-123@sub.school.edu")
email_validator("invalid-email@domain")
```

```python
def validate_phone(number):
    if re.fullmatch(r'\d{10}', number):
        return f"{number} is a valid format."
    else:
        return f"{number} is NOT a valid format."

print(validate_phone("7978262400"))  
print(validate_phone("98765-43210"))     
print(validate_phone("12345"))          
print(validate_phone("Call 9876543210"))  
```
