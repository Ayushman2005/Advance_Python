```python
import re

def check_password(password):
    if len(password) < 8:
        return "Password must be at least 8 characters."

    if not re.search(r"[A-Z]", password):
        return "Missing a capital letter (A-Z)."

    if not re.search(r"[a-z]", password):
        return "Missing a lowercase letter (a-z)."

    if not re.search(r"\d", password):
        return "Missing a number (0-9)."

    if re.search(r"[A-Za-z0-9]", password):
        return "Missing a symbol"

    return "Strong password! You're good to go."

test_pwd = input("Enter password: ")
print(test_pwd , check_password(test_pwd))
```

```python
import re

def clean_sentence(text):
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text)
    
    return text

sentence = "Hello!!!   My NAme @is ... Ayushman202024x??"
print(clean_sentence(sentence)) 
```

```python
import re
from PyPDF2 import PdfReader
reader = PdfReader("example.pdf")
data=""
for page in reader.pages:
    data += page.extract_text()
words = re.findall(r'\b[A-Za-z]+\b', data)
numbers = re.findall(r'\d+', data)
emails = re.findall(r'[\w\.-]+@[\w\.-]+\.\w+', data)
print("Extracted Text:\n", data)

print("\nWords Found:")
print(words)

print("\nNumbers Found:")
print(numbers)

print("\nEmails Found:")
print(emails)
```
